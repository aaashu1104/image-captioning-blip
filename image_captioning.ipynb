```python
# Step 1: Install required packages
!pip install transformers pillow torchvision

# Step 2: Import Libraries
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests
import torch
from google.colab import files

# Step 3: Upload Your Own Image
uploaded = files.upload()  # This will prompt you to upload an image
image_path = list(uploaded.keys())[0]
image = Image.open(image_path).convert('RGB')

# Step 4: Load Pretrained Model (BLIP)
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to("cuda" if torch.cuda.is_available() else "cpu")

# Step 5: Generate Caption
inputs = processor(image, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")
out = model.generate(**inputs)
caption = processor.decode(out[0], skip_special_tokens=True)

# Step 6: Show Result
print(f"🖼️ Generated Caption: {caption}")
image.show()
```
